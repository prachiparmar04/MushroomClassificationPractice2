 #### **PROBLEM STATEMENT:**
 `This data set includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family (pp. 500-525). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; 
 

#### Attribute Information:
    - TARGET CLASS: (class: edible=e =0, poisonous=p = 1)

    - cap-shape: bell=b,conical=c,convex=x,flat=f, knobbed=k,sunken=s

    - cap-surface: fibrous=f,grooves=g,scaly=y,smooth=s

    - cap-color: brown=n,buff=b,cinnamon=c,gray=g,green=r,pink=p,purple=u,red=e,white=w,yellow=y

    - bruises: bruises=t-1,no=f-0

    - odor: almond=a,anise=l,creosote=c,fishy=y,foul=f,musty=m,none=n,pungent=p,spicy=s

    - gill-attachment: attached=a,descending=d,free=f,notched=n

    - gill-spacing: close=c,crowded=w,distant=d

    - gill-size: broad=b,narrow=n

    - gill-color: black=k,brown=n,buff=b,chocolate=h,gray=g, green=r,orange=o,pink=p,purple=u,red=e,white=w,yellow=y

    - stalk-shape: enlarging=e,tapering=t

    - stalk-root: bulbous=b,club=c,cup=u,equal=e,rhizomorphs=z,rooted=r,missing=?

    - stalk-surface-above-ring: fibrous=f,scaly=y,silky=k,smooth=s

    - stalk-surface-below-ring: fibrous=f,scaly=y,silky=k,smooth=s

    - stalk-color-above-ring: brown=n,buff=b,cinnamon=c,gray=g,orange=o,pink=p,red=e,white=w,yellow=y

    - stalk-color-below-ring: brown=n,buff=b,cinnamon=c,gray=g,orange=o,pink=p,red=e,white=w,yellow=y

    - veil-type: partial=p,universal=u

    - veil-color: brown=n0,orange=o1,white=w3,yellow=y4

    - ring-number: none=n,one=o,two=t

    - ring-type: ,evanescent=e,flaring=f,large=l,none=n,pendant=p,

    - spore-print-color: black=k,brown=n,buff=b,chocolate=h,green=r,orange=o,purple=u,white=w,yellow=y

    - population: abundant=a,clustered=c,numerous=n,scattered=s,several=v,solitary=y

    - habitat: grasses=g,leaves=l,meadows=m,paths=p,urban=u,waste=w,woods=d

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,roc_auc_score,roc_curve
```

##### 1. Choosing Random State 

```python

randomstate = 0
max_score = 0
lr = LogisticRegression()
for i in range(1,46):
    x_train,x_test,y_train,y_test = train_test_split(labeled_data,y,test_size=.30,random_state = i)
    lr.fit(x_train,y_train)
    pred = lr.predict(x_test)
    score = accuracy_score(y_test,pred)    
    if(score>max_score):
        max_score = score
        randomstate  = i
print( 'Random State=',randomstate,' and R2 Score of LR MODEL : ',max_score*100)
```

#####  **Observations**
    - Random State Choosen is 32 which gives highest accuracy on a model

    - Lets Check out accuracy w.r.t to different Models

##### 2. Lets Model different Models and Evaluate them

```python
x_train,x_test,y_train,y_test = train_test_split(labeled_data,y,test_size=.30,random_state = 32)
```

```python
x_train.shape, x_test.shape
```

##### 5.2.1: KNN

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train,y_train)
```

```python
knn_acc = 0
k = 0
error = []
for i in range(1,80):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    score = accuracy_score(y_test,pred_i)
    error.append(np.mean(pred_i != y_test))
    if(score> knn_acc):
        knn_acc = score
        k = i
print('KNN Accuracy :', knn_acc, ' and its K value is :',k)
```

```python
plt.figure(figsize=(15,10))
plt.title('ErrorRate VS K-Value')
plt.plot(range(1,80),error,marker='*',linestyle='dashed')
```

```python
accuracy_scores = []
cv_scores = []
```
